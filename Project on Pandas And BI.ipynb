```python
import pandas as pd
```

```python
import json
```

```python
#match Summary
```

```python
with open("E:/t20_json_files/t20_wc_match_results.json") as f:
    data = json.load(f)
    data[0]['matchSummary']
```

```python
len(data)
```

```python
 data[0]['matchSummary']
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis =1 , inplace = True)
```

```python
df_match.head()
```

```python
#to add two tables
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
    match_ids_dict
```

```python
match_ids_dict
```

```python
df_match.to_csv('E:/ASHISH/Cric T20 Power BI project/T20/dim_match_summary.csv', index = False)
```

```python
#Batting Summary
```

```python
with open("E:/t20_json_files/t20_wc_batting_summary.json") as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
        
    
```

```python
all_records
```

```python
df_batting =  pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting.tail()
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not_out')
df_batting.head(11)
```

```python
df_batting.drop(columns=['dismissal'], inplace = True)
df_batting.head(11)
```

```python
#replacing special characters
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
#to add two tables
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
match_ids_dict['Namibia Vs Sri Lanka']
```

```python
#converting into csv
```

```python
df_batting.to_csv('E:/ASHISH/Cric T20 Power BI project/T20/fact_batting_summary.csv', index = False) 
```
